```python
from sklearn.datasets import make_regression
```

```python
import numpy as np
```

```python
X,y =make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
# lets simpy apply ols
# to find the value of slope and intercept
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg =LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')


```

```python
# Lets apply Gradient Descent assuming slope is constant m = 78.35
# and let's assume the starting value for intercept b = 0
```

```python
y_pred =((78.35)*X +0).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='#00a65a',label='b = 0')
plt.legend()
plt.show()
```

```python
m=78.35
b=0
loss_slope =-2*np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
#"Ravel" ka matlab hai kisi cheez ko seedha karna
#ya uske layers ko kholna, jaisa ke aapne code mein dekha,
#jahan X.ravel() ne array ko seedha, yani 1D, bana diya.

```

```python
# lets take learning rate =0.1
lr=0.1
step_size =lr*loss_slope
step_size
```

```python
# calculating the new_intercept
b= b-step_size
b
```

```python
y_pred1 =(78.35*X +b).reshape(4)
y_pred1
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1,color='#00a65a',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```

```python

b = -100
m = 78.35
lr = 0.01

epochs = 100

for i in range(epochs):
  loss_slope = -2 * np.sum(y - m*X.ravel() - b)
  b = b - (lr * loss_slope)

  y_pred = m * X + b

  plt.plot(X,y_pred)

plt.scatter(X,y)

```

It seems there is a typo in the code. The method `revel()` is not a standard NumPy method. The intended method is likely `ravel()`, which returns a flattened view of an array. This means it converts a multi-dimensional array into a 1D array.
